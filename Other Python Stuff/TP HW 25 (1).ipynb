```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 0.007
a = 0.00005

Ce = lambda T: g*T
Cv = lambda T: a*T**3
T = np.arange(0,20)

fig = plt.figure()
ax = fig.add_axes([0.15,0.1,0.8,0.8])
ax.plot(T,Ce(T), label='$Electronic$')
ax.plot(T,Cv(T), label='$Vibration$')
ax.legend();
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n1= 1
n2= 1.5
theta1= []

for i in range (91):
    theta1.append(np.deg2rad(i))

theta2= np.deg2rad(np.arcsin(n1/n2*np.sin(theta1)))

R= ((n2*np.cos(theta1)-n1*np.cos(theta2))/(n1*np.cos(theta2)+ n2*np.cos(theta1)))
T= ((2*n1*np.cos(theta1))/(n2*np.cos(theta1)+ n1*np.cos(theta2)))
```

```python
plt.title( "Reflexivity as a Function of the Incident Angle")
plt.scatter(theta1,R)
plt.xlabel("Radians")
plt.ylabel("Reflexivity");
```

```python
plt.title("Transmissivity as a Function of the Incidnet Angle")
plt.scatter(theta1, T)
plt.xlabel("Radians")
plt.ylabel("Transmissivity");
```
